# Some basics 

-  Plot a the sine function from 0 to 6 (inclusinve) using 50 points.
-  Plot the line again in red and with diamond shaped markers of size 5.
-  Now plot the line, but include a legend.

*Helpful functions*: ``plt.plot``, ``plt.legend``, ``plt.grid``

## Sol

```python
x = np.linspace(0, 6, 50)
plt.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x), 'rd', markersize = 5)
```

```python
plt.plot(x, np.sin(x), '-', label = 'sin(x)')
plt.legend()
```

## Sol

```python
x = np.linspace(0,1,100)

fig, (ax1, ax2) = plt.subplots(nrows = 2, ncols =1, sharex = True, squeeze=True)

ax1.plot(x, np.sin(2*np.pi*x)*np.exp(-5*x), 'b-', linewidth=5, label = r"$\sin\left(2\pi t\right)e^{-5t}$")
ax1.plot(x, np.zeros_like(x), 'k-', linewidth = 1)
ax1.fill_between(x, np.sin(2*np.pi*x)*np.exp(-5*x), color = 'r')
ax1.legend()
ax1.set_ylabel('position')

ax2.plot(x, np.sin(2*np.pi*x), 'b-', linewidth =5, label = r'$\sin\left(2\pi t\right)$')
ax2.plot(x, np.cos(2*np.pi*x), 'b-', linewidth =5, label = r'$\cos\left(2\pi t\right)$')

ax2.fill_between(x, np.sin(2*np.pi*x), np.cos(2*np.pi*x), color = 'g')
ax2.legend()
ax2.set_xlabel('time')
```
